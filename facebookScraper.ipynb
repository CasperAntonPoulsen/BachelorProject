```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
import bs4
from bs4 import BeautifulSoup as soup
import time

opts = Options()
opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246")

driver = webdriver.Chrome("chromedriver.exe")

driver.get("https://www.facebook.com/DRNyheder")

actions = ActionChains(driver)

for i in range(100):
    time.sleep(4)
    element = driver.find_elements_by_link_text("DR Nyheder")[-1]

    actions.move_to_element(element).perform()

```
