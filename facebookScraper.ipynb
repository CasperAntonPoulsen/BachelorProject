```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
import bs4
from bs4 import BeautifulSoup as soup
import time
import json


opts = Options()
opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246")

driver = webdriver.Chrome("chromedriver.exe")

driver.get("https://www.facebook.com/DRNyheder")


scrollElementIntoMiddle = "var viewPortHeight = Math.max(document.documentElement.clientHeight, window.innerHeight || 0);" + "var elementTop = arguments[0].getBoundingClientRect().top;" + "window.scrollBy(0, elementTop-(viewPortHeight/2));"
```

```python
comments = []
for i in range(0,500000):
    actions = ActionChains(driver)

    post = driver.find_elements(By.XPATH, "//div[@aria-posinset]")[i]

    time.sleep(1)
    try:
        mest = post.find_element(By.XPATH, ".//*[contains(text(), 'Mest relevante')]")
    except:
        continue
    driver.execute_script(scrollElementIntoMiddle, mest)

    time.sleep(1)
    ActionChains(driver).move_to_element(mest).click(mest).perform()
    
    time.sleep(1)
    
    driver.find_elements(By.XPATH, "//div[@role='menuitem']")[-1].click()

    time.sleep(1)
    
    while True:
        flere = post.find_elements(By.XPATH, ".//*[contains(text(), 'flere kommentarer')]")
        if flere != []:
            driver.execute_script(scrollElementIntoMiddle, flere[0])

            ActionChains(driver).move_to_element(flere[0]).click(flere[0]).perform()
            time.sleep(1)
        mere = post.find_elements(By.XPATH, ".//*[contains(text(), ' kommentarer mere')]")
        if mere != []:
            driver.execute_script(scrollElementIntoMiddle, mere[0])

            ActionChains(driver).move_to_element(mere[0]).click(mere[0]).perform()
            time.sleep(1)
                
        merer = post.find_elements(By.XPATH, ".//*[contains(text(), ' kommentar mere')]")
        if merer != []:
            driver.execute_script(scrollElementIntoMiddle, merer[0])

            ActionChains(driver).move_to_element(merer[0]).click(merer[0]).perform()
            time.sleep(1)
        if flere == [] and mere == [] and merer == []:
            break

    post = driver.find_elements(By.XPATH, "//div[@aria-posinset]")[i]
    for mere in post.find_elements(By.XPATH, ".//*[contains(text(), ' svar')]"):
        driver.execute_script(scrollElementIntoMiddle, mere)
        time.sleep(0.25)
        ActionChains(driver).move_to_element_with_offset(mere, -200, 0).click(mere).perform()


    post = driver.find_elements(By.XPATH, "//div[@aria-posinset]")[i]
    for mere in post.find_elements(By.XPATH, ".//*[contains(text(), 'Se mere')]"):
        driver.execute_script(scrollElementIntoMiddle, mere)
        time.sleep(0.25)
        ActionChains(driver).move_to_element_with_offset(mere, -200, 0).click(mere).perform()


    post = driver.find_elements(By.XPATH, "//div[@aria-posinset]")[i]
    for comment in post.find_elements(By.XPATH, ".//li/div[@aria-label]"):

        if comment == []:
            continue


        maybeComment = comment.find_elements(By.XPATH, ".//span[@dir='auto']")

        if len(maybeComment) < 2:
            continue
            

        with open("drNyheder.json", "a") as file:
            file.write(json.dumps({"commenter":maybeComment[0].text,"comment": maybeComment[1].text}) + "\n")
                #comments.append({"commenter":maybeComment[j].text,"comment": maybeComment[j+1].text})


        #commentBreakdownSplit = commentBreakdown[0].text.split("\n")
    

        #comments.append({"comment": commentBreakdownSplit[1], "commenter": commentBreakdownSplit[0]})

```
